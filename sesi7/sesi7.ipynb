```python
# Line Chart
df = pd.DataFrame({
    "year":[2010, 2011, 2012, 2013, 2014, 2015, 2016],
    "subscriber": [1000, 2000, 2300, 4000, 7000, 14000, 10000]
})
```

```python
df.head()
```

```python
plt.plot(df["year"], df["subscriber"])
plt.xlabel("year")
plt.ylabel("subscriber")
plt.title("Subscriber Progress")
plt.show()
```

```python
# Bar Chart
df = pd.DataFrame({
    "provinsi": ["jakarta", "jawa tengah", "jawa timur", "sumut", "bengkulu"],
    "pengangguran":[1000, 1200, 900, 1300, 800]
})
```

```python
plt.bar(df["provinsi"], df["pengangguran"])
plt.xlabel("provinsi")
plt.ylabel("pengangguran")
plt.title("Perbandingan Pengangguran Tahun 1800")
plt.show()
```

```python
# Area Chart
df = pd.DataFrame({
    "year":[2010, 2011, 2012, 2013, 2014, 2015, 2016],
    "subscriber": [1000, 2000, 2300, 4000, 7000, 14000, 10000],
    "new_subscriber": [1000, 100, 300, 2000, 3000, 10000, 1000],
    "unsubscriber": [0, 200, 400, 200, 1000, 100, 5000],
})
# subscriber
plt.fill_between(df["year"], df["subscriber"], color='skyblue')
plt.plot(df["year"],df["subscriber"], label="subscriber")

plt.fill_between(df["year"], df["new_subscriber"], color='orange')
plt.plot(df["year"],df["new_subscriber"], label="new_subscriber")

plt.fill_between(df["year"], df["unsubscriber"], color='green')
plt.plot(df["year"],df["unsubscriber"], label="unsubscriber")

plt.xlabel("year")
plt.ylabel("amount")
plt.legend() 
plt.show()
```

```python
# Pie Chart
# menunjukkan porsi perbandingan
df = pd.DataFrame({
    "provinsi": ["jakarta", "jawa tengah", "jawa timur", "sumut", "bengkulu"],
    "pengangguran":[1000, 1200, 900, 1300, 800]
})
plt.pie(df["pengangguran"], labels=df["provinsi"], autopct='%1.1f%%')
plt.title("Persentase Pengangguran di Indonesia Tahun 1800")
plt.show()
```

```python
# Histogram
# persebaran frekuensi kemunculan data
df = pd.DataFrame({
    "nilai":[55, 60, 70, 75, 80, 82, 85, 90, 95, 100]
})
plt.hist(df["nilai"], bins=5)
plt.title("Persebaran Nilai")
plt.show()
```

```python
# Box Plot
df = pd.DataFrame(
    {'jenis': ['A', 'A', 'A', 'B', 'B', 'C', 'C', 'C', 'D', 'D', 'D', 'D'],
        'nilai': [20, 25, 30, 10, 15, 5, 7, 10, 30, 35, 40, 45]}
)
plt.boxplot(
   [
    df[df["jenis"] == "A"]["nilai"],
    df[df["jenis"] == "B"]["nilai"],
    df[df["jenis"] == "C"]["nilai"],
    df[df["jenis"] == "D"]["nilai"]
   ]
)

plt.xticks([1, 2, 3, 4], ["A", "B", "C", "D"])

plt.xlabel("jenis")
plt.xlabel("nilai")
plt.show()
```

```python
# Scatter Plot
df = pd.DataFrame({
    "x" : [1, 3, 5, 6, 7, 8, 9, 10],
    "y" : [4, 6, 8, 10, 11, 14, 15, 20],
})
plt.scatter(df["x"], df["y"])
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# data set canada
df_canada = pd.read_excel("./canada.xlsx", 
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2)
df_canada.head(100)
```

```python
#  Plot pertumbuhan imigran asia (Line chart)
years = []
for col in df_canada.columns:
    if isinstance(col, int):
        years.append(col)
years
```

```python
# Plot immigration for Indonesia
plt.plot(years, df_canada.loc["Indonesia", years])
plt.xlabel("year")
plt.ylabel("amount")
plt.title("Data imigrasi Indonesia")
plt.show()
```

```python
df_asia = df_canada[df_canada["AreaName"] == "Asia"][years]
df_asia.index
```

```python
# plot semua immigration from Asia
for country in df_asia.index:
    plt.plot(years, df_asia.loc[country], label=country)
plt.xlabel("year")
plt.ylabel("amount")
plt.title("Data imigrasi Asia")
# plt.legend()
plt.show()
```

```python
# get data from India and China
plt.plot(years, df_asia.loc["India"], label="India")
plt.plot(years, df_asia.loc["China"], label="China")
plt.xlabel("year")
plt.ylabel("amount")
plt.title("Data imigrasi Asia")
plt.legend()
plt.show()
```

```python
df_asia.transpose()
```

```python
df_asia["Total"] = df_asia.sum()
df_asia
```

```python
df_asia[years].sum().to_frame()
```

```python
# Plot in line, total all Asia Immigrant every year

plt.plot(years, df_asia[years].sum().to_frame()[0])
plt.xlabel("year")
plt.ylabel("amount")
plt.title("Data imigrasi Asia")
plt.show()
```

```python
# Buatlah Percentage per AreaName
df_canada["AreaName"].unique()
df_canada.head()
```

```python
years
```

```python
filtered_column = ["AreaName"] + years
```

```python
filtered_column
```

```python
df_per_area = df_canada[filtered_column]
df_per_area.set_index("AreaName", inplace=True)
df_per_area.head()
```

```python
df_per_area_grouped = df_per_area.groupby("AreaName").sum()
df_per_area_grouped.head()
df_area = df_per_area_grouped.transpose().sum().to_frame()
df_area["Total"] = df_area[0]
df_area
```

```python
p, t = plt.pie(df_area["Total"])
plt.title("Immigration to Canada [1980 - 2013] per Area")
plt.legend(p, df_area.index)
plt.axis("equal")
plt.show()
```

```python
df_area["Total"].plot(
    kind="pie",
    autopct='%1.1f%%',
)
plt.title("Immigration to Canada [1980 - 2013] per Area")
plt.legend()
plt.axis("equal")
plt.show()
```
