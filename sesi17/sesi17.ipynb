```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd

# breast cancer dataset
from sklearn.datasets import load_breast_cancer
```

```python
# load the breast cancer dataset
breast_data = load_breast_cancer().data
breast_data.shape
```

```python
breast_data
```

```python
breast_label = load_breast_cancer().target
breast_label.shape
```

```python
breast_label
```

```python
# reshape the breast_label to concatenate it with the breast_data
breast_label = breast_label.reshape(569,1)  

# concatenate the data and the label
breast_clean = np.concatenate([breast_data, breast_label], axis=1)  
breast_clean
```

```python
# transform to dataframe
breast_pd = pd.DataFrame(breast_clean)
breast_pd.head()

# features name 
features = load_breast_cancer().feature_names
features_with_label = np.append(features, 'label')
# add column to dataframe
breast_pd.columns = features_with_label
breast_pd.head()
```

```python
# replace label 0 with 'Benign' and 1 with 'Malignant'
breast_pd['label'].replace(0, 'Benign',inplace=True)    
breast_pd['label'].replace(1, 'Malignant',inplace=True)
breast_pd.head()
```

```python
# PCA normalization
from sklearn.preprocessing import StandardScaler
x = breast_pd.loc[:, features].values
x = StandardScaler().fit_transform(x) # normalizing the features
# print data before and after normalization
pd.DataFrame(x, columns = features).head()
```

```python
# create PCA using sklearn
from sklearn.decomposition import PCA
breast_pca = PCA(n_components=2)
```

```python
# pca model
principalComponents = breast_pca.fit_transform(x)
principalComponents
```

```python
# create dataframe
principal_breast = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2'])
principal_breast.head()

# summarize the explained variance ratio
print('Explained variation per principal component: {}'.format(breast_pca.explained_variance_ratio_))

# conclusion: 
# the first principal component contains 44.27% of the variance and 
# the second principal component contains 19.38% of the variance. 
# Together, the two components contain 63.65% of the information.
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
plt.scatter(principal_breast['PC1'], principal_breast['PC2'], c=breast_pd['label'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Visualization')
plt.show()
```

```python
from sklearn.cluster import KMeans

# K-means clustering
# Specify the number of clusters
k = 2

# Create an instance of KMeans
kmeans = KMeans(n_clusters=k)

# Fit the data to the KMeans model
kmeans.fit(principalComponents)

# Get the cluster labels
# label = predicted cluster
labels = kmeans.labels_

print(labels)

# Plot the clusters
plt.scatter(principalComponents[:, 0], principalComponents[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', color='red')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-Means Clustering')
plt.show()
```

```python
# evaluate kmeans with actual
np.mean(labels != breast_pd['label'])
```
