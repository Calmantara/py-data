```python
# OWN_OCCUPIED => jika dia ada nilainya, maka dianggap sebagai Y. jika NAN, maka dianggap N
id = 0
for val in df["OWN_OCCUPIED"]:
    # check apakah setiap row string
    try:
        # if not error, meaning type is int
        val_int = int(val)
        df.loc[id, "OWN_OCCUPIED"] = "N"
        if val_int > 0:
            df.loc[id, "OWN_OCCUPIED"] = "Y"
    except ValueError:
        pass
    id += 1
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/property_data.csv', na_values=["n/a", "--", "na"])

def masking_occupied(val):
    try:
        # if not error, meaning type is int
        val_int = int(val)
        if val_int > 0:
            return "Y"
        return "N"
    except ValueError:
      return val  

df["OWN_OCCUPIED"] = df["OWN_OCCUPIED"].apply(masking_occupied)
```

```python
df
```

```python
# PID => jika PID NaN, maka boleh di remove columnnya
df = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/property_data.csv', na_values=["n/a", "--", "na"])
df.dropna(subset=["PID"], inplace=True)
df
```

```python
# SQ_FT < 1000, maka boleh kita drop
df = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/property_data.csv', na_values=["n/a", "--", "na"])
for idx in df.index:
    if df.loc[idx, "SQ_FT"] < 1000:
        df.loc[idx, "SQ_FT"] = np.nan
df.dropna(subset=["SQ_FT"], inplace=True)
```
